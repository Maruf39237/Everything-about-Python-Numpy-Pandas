# ***recursion***

```python
def display(n):
    if n == 0 : return 0
    print(n)
    display(n-1)

def factorial(n) :
    if(n == 1) : return 1
    return n * factorial(n-1)



print("5 to 1 : ")
display(5)
print("fatorial of 10 is : ", factorial(10))
```

```python
# binary serach

array = [1, 1, 2, 3, 2, 4, 5, 6, 7, 5, 6, 7, 8, 99, -1, 2]

def binary_search(item):
    l = 0
    r = len(array)-1
    mid = None

    while l <= r :
        mid = int((l + r) / 2)
        if array[mid] == item :
            print("item found...")
            return
        elif array[mid] > item :
            r = mid - 1
        else :
            l = mid + 1

    print("item not found...")




array.sort()
print(array)

item = int(input('enter the value you want to search : '))
binary_search(item)
```

```python
# prefix sum

arr = [   1, 1, 2, 3, 2, 4]
#      0  1  2  3  4  5  6
prefix = [0] * (len(arr)+1)


for i in range(1, len(arr)+1):
    prefix[i] = prefix[i-1] + arr[i-1]

print("the prefix sum series : ", prefix)

l = int(input('enter the value of l : '))
r = int(input('enter the value of r : '))

print(prefix[r]-prefix[l-1])
```
