# 42. List Duplicates Removal: Write a Python program to remove duplicates from a given list while preserving the order of the elements


```python
lst = [5, 5, 4, 3, 4, 3, 2, 1, 2, 1, -1, 2, -1, 2]
inf = 1e9
nlst = []

for i in range(0, len(lst)) :
    if lst[i] != inf :
        nlst.append(lst[i])
        for j in range(i+1, len(lst)) :
            if lst[i] == lst[j] :
                lst[j] = inf

print(nlst)
```

# 108. Math Operations: Write a Python function called `math_operations` that takes three numbers and a string representing an operation (‘add’, ‘subtract’, ‘multiply’, or ‘divide’). The function should return the result of the specified operation on the three numbers. Implement the math operations as nested functions.



```python
def math_operations(a, b, c, op):
    def add():
        return a + b + c

    def subtract():
        return a - b - c

    def multiply():
        return a * b * c

    def division():
        try:
            return a / b / c
        except ZeroDivisionError:
            return "error: division by zero"

    dct = {
        'add': add,
        'subtract': subtract,
        'multiply': multiply,
        'division': division
    }

    if op in dct:
        return dct[op]()
    else:
        return "error: invalid operation"

print(math_operations(2, 3, 4, 'add'))
print(math_operations(4, 3, 2, 'subtract'))
print(math_operations(2, 3, 4, 'multiply'))
print(math_operations(10, 5, 2, 'division'))
print(math_operations(10, 5, 0, 'division'))
print(math_operations(2, 3, 4, 'invalid'))
```

```python
def temperature_converter(temp, s) :
    def celc() :
        return ((temp - 32) / (9 / 5))
    def farh() :
        return ((temp * 9 / 5) + 32)

    dct = {
        'celc' : celc,
        'farh' : farh
    }

    if s in dct :
        return dct[s]()

print(temperature_converter(80, 'celc'))
print(temperature_converter(80, 'farh'))
```

# 114. Binary Search: Write a recursive Python function called `binary_search` that takes a sorted list and a target value as input and returns the index of the target value in the list using binary search. If the target value is not in the list, return -1.



```python
item = int(input())
lst = [9, 8, 7, 6, 5, 4, 3, 2, 1]
lst.sort()

l = 0
r = len(lst)-1
f = False
while(l <= r) :
    mid = int((l + r) / 2.0)
    if lst[mid] == item :
        f = True
        break
    elif item < lst[mid] :
        r = mid - 1
    else :
        l = mid + 1

if f == True :
    print("item found")
else :
    print('item not found')
```

```python
def binary_search(lst, item, l, r) :
    if l > r :
        return -1

    mid = int((l + r) / 2.0)
    if item == lst[mid] :
        return mid
    elif item < lst[mid] :
        return binary_search(lst, item, l, mid - 1)
    elif item > lst[mid] :
        return binary_search(lst, item, mid + 1, r)


lst = [9, 8, 7, 6, 5, 4, 3, 2, 1]
lst.sort()
binary_search(lst, 100, 0, len(lst)-1)
```

# 115. Power Calculation: Write a recursive Python function called `power` that takes two positive integers, base and exponent, as input and returns the value of base raised to the exponent.



```python
def power(b, e) :
    if e == 0 :
        return 1
    else :
        return b * power(b, e - 1)

power(4, 2)
```
